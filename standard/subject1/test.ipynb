```python
import pandas as pd

df = pd.read_csv("C:/msys64/home/for/10th/standard/flight_data_homework.csv")

Des_Pr = df.groupby("Destination")["Price"].sum()
des_pr_mean = df.groupby("Destination")["Price"].mean()
des_pr_median = df.groupby("Destination")["Price"].median()

mean2 = round(des_pr_mean, 1)
median2 = round(des_pr_median, 1)

# destination기준 price 평균, 중앙값
print(mean2)
print(median2)

# airline, Total_stops기준 Route 중복없이 count
df2 = df.groupby(["Airline", "Total_Stops"])["Route"].nunique().reset_index()
df2
```

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv("C:/msys64/home/for/10th/standard/flight_data_homework.csv")

# 피벗테이블, 출발지 도착지 기준 count airline, desc
pivot = pd.pivot_table(df, index=['Source', 'Destination'], values = 'Airline', aggfunc='count').sort_values('Airline', ascending = False).reset_index()


# 필터링
filter = df.loc[(df["Airline"] == "Air India") & (df["Price"] >= 7000)]



display(pivot)
display(filter)
```
