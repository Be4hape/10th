```python
#정답코드

import pandas as pd
from dateutil.rrule import rrule, DAILY, WEEKLY, MONTHLY, YEARLY, MO, TU, WE, TH, FR, SA, SU

df = pd.read_csv("C:/msys64/home/for/10th/standard/flight_data_homework.csv")
df["Date_of_Journey"] = pd.to_datetime(df["Date_of_Journey"], infer_datetime_format=True)

start_date = df["Date_of_Journey"].min()
end_date = df["Date_of_Journey"].max()

n1 = list(rrule(WEEKLY, dtstart=start_date, until=end_date, byweekday=WE))
n2 = df[df["Date_of_Journey"].isin(n1)]['Price'].mean()
n2
```

```python
#람다 안에서 인라인으로 조건문을 만드는 방법
# (1)만족값 (2)if (3)조건 (4)못했을때 결과값

import pandas as pd
from IPython.display import display
import datetime as dt

df = pd.read_csv("C:/msys64/home/for/10th/standard/flight_data_homework.csv")

display(df.T)

# df["Dep_Time"].info
df["Dep_Time"] = pd.to_datetime(df["Dep_Time"], format="%H:%M")
# df["Dep_Time"].info

df["time_sep"] = df["Dep_Time"].dt.hour.apply(
    lambda x : (
        "mor" if 5 <= x < 12
        else "aft" if 12 <= x < 18
        else "din" if 18 <= x < 24
        else "night"
    )    
)
df.groupby("time_sep")["Airline"].count()
```
