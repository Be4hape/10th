```python
import pandas as pd
from IPython.display import display
import datetime as dt

df = pd.read_csv("C:/msys64/home/for/10th/standard/flight_data_homework.csv")

display(df.T)

# df["Dep_Time"].info
df["Dep_Time"] = pd.to_datetime(df["Dep_Time"], format="%H:%M")
# df["Dep_Time"].info

df["time_sep"] = df["Dep_Time"].dt.hour.apply(
    lambda x : (
        "mor" if 5 <= x < 12
        else "aft" if 12 <= x < 18
        else "din" if 18 <= x < 24
        else "night"
    )    
)
df.groupby("time_sep")["Airline"].count()
```
