```python
# pandas 라이브러리를 활용한 csv 파일 읽기 
df = pd.read_csv("product_details.csv") # product_details.csv
df2 = pd.read_csv("customer_details.csv") # customer_details.csv
df3 = pd.read_csv("E-commerece sales data 2024.csv") # E-commerece sales data 2024.csv
```

```python
# pivot table
df2 = pd.pivot_table(df2, index='Payment Method', columns=['Size','Season'], values='Customer ID', aggfunc='count')
```

```python
# melt 를 통해 컬럼을 열로 바꿔주기 
# 인덱스를 컬럼으로 변환(인덱스 재정렬)
df2 = df2.reset_index()  
```

```python
# id_vars : 기준이 될 열 ( 그대로 유지할 열 )
# value_vars : 기준열에 대한 하위 카테고리를 나열할 열을 선택 ( 병합할 열 )
# var_name : 카테고리들이 나열된 열의 이름 설정 ( 병합 후 표현할 열 이름 )
# value_name : 카테고리들의 값이 나열될 열의 이름 설정 ( 기본값: value )
df2.melt(id_vars=['Payment Method'], var_name=['Size', 'Season'], value_name='Customer ID')
```

```python
## melt 연습3
# id_vars : 기준이 될 열 ( 그대로 유지할 열 )
# value_vars : 기준열에 대한 하위 카테고리를 나열할 열을 선택 ( 병합할 열 )
# var_name : 카테고리들이 나열된 열의 이름 설정 ( 병합 후 표현할 열 이름 )
# value_name : 카테고리들의 값이 나열될 열의 이름 설정 ( 기본값: value )
# location 열을 기준으로 구매주기 컬럼을 풀어주어 기존 컬럼에 있던 값이 열로 배치되었습니다.
df2.melt(id_vars=['Location'], value_vars=['Frequency of Purchases'])
```

```python
# stack 
df2 = pd.read_csv("customer_details.csv") # customer_details.csv

# 필요한 컬럼 선택
df22 = df2[['Category', 'Season', 'Purchase Amount (USD)', 'Review Rating']].copy()

# pd.MultiIndex from tuples 를 통해, 멀티인덱스 설정
df22.columns = pd.MultiIndex.from_tuples([('Purchase', 'Amount'), ('Purchase', 'Review'),
                                              ('Info', 'Category'), ('Info', 'Season')])

df22
```

```python
# stack(level=0) 적용
df_stacked_0 = df22.stack(level=0)

# stack(level=1) 적용
df_stacked_1 = df22.stack(level=1)

# 결과 비교 
display(df22, df_stacked_0, df_stacked_1)
```

```python
# unstack
# unstack(level=-1) 적용. 기본값
df_unstacked_0 = df_stacked_0.unstack(-1).reset_index()
# 결과 비교 
display(df22, df_stacked_0, df_unstacked_0)
```
