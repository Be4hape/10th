```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:/msys64/home/for/10th/standard/LinearBNB/2025_Airbnb_NYC_listings.csv")
```

```python
# 필요한 전체 컬럼 선정
df = df[['neighbourhood_group_cleansed', 'room_type', 'accommodates', 'bathrooms', 'neighbourhood_cleansed',
    'latitude', 'longitude',
    'bedrooms', 'beds', 'amenities', 'minimum_nights']].copy()

df.T
```

```python
def count_amenities(x):
    if isinstance(x, str):
        clean_x = x.replace('[', '').replace(']', '').replace('"', '').replace("'", "")
        if clean_x == "": return 0
        return len(clean_x.split(','))
    elif isinstance(x, list):
        return len(x)
    return 0

df['amenities_count'] = df['amenities'].apply(count_amenities)

df.T
```

```python
# (1.5개 같은 화장실을 2.0으로 만듭니다) # 상한선(Capping) 적용
df['bathrooms'] = np.ceil(df['bathrooms'])
df['bathrooms_clipped'] = df['bathrooms'].clip(upper=5.0)

df.T
```

```python
df['minimum_nights_cleansed'] = df['minimum_nights'].clip(lower=1, upper=30)

df.T
```

```python
from haversine import haversine

center = (40.7580, -73.9855)
df['dist_from_center(km)'] = df.apply(lambda x: haversine((x['latitude'], x['longitude']), center, unit='km'), axis=1)

df.T
```

```python
import ast
import re

# 편의시설 파싱 함수
def clean_amenities(x):
    if pd.isna(x): return []
    x = x.replace('{', '[').replace('}', ']')
    try:
        return [i.strip().lower() for i in ast.literal_eval(x)]
    except:
        return [i.strip().lower() for i in re.sub(r'[\[\]\"\'\{\}]', '', x).split(',') if i.strip()]

df['amenities_list'] = df['amenities'].apply(clean_amenities)


# 카테고리 정의 (키워드 기반)
categories = {
    'luxury_score': ['elevator', 'gym', 'pool'],
    'service_score': ['washer', 'coffee maker', 'self check-in', 'pet'],
    'longterm_score': ['shampoo', 'dryer', 'iron'],
    'essential_score' : ['air conditioning', 'essentials']
}

# 각 카테고리별 점수 계산 (보유한 항목의 개수)
for cat_name, keywords in categories.items():
    df[cat_name] = df['amenities_list'].apply(lambda x: sum(1 for item in x if any(k in item for k in keywords)))

df.T
```

```python
df = df.drop(columns=['amenities_list', 'longterm_score', 'essential_score'])

df.T
```
